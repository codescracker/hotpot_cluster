```python
res = []
for i in range(0,len(data_list)):
    temp = {}
    group = None
    for index in range(0,len(cluster)):
        if data_list[i] in cluster[index]:
            group = index
    uid = record_sequence[i]
    temp['group'] = group
    temp['uid'] = uid
    for j in range(0,len(feature_sequence)):
        temp[feature_sequence[j]] = data_list[i][j]
    res.append(temp)
    
```

```python
for restaurant in res:
    for detail in d:
        if restaurant['uid'] == detail['uid']:
            restaurant['lat'] = detail['location']['lat']
            restaurant['lng'] = detail['location']['lng']
            if 'overall_rating' in detail['detail_info']:
                restaurant['overall_rating'] = detail['detail_info']['overall_rating']
            else:
                restaurant['overall_rating']= 'Unknown'
            if 'price' in detail['detail_info']:
                restaurant['price'] = detail['detail_info']['price']
            else:
                restaurant['price'] = 'Unknown'
```
